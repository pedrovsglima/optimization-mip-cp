```python
import pandas as pd
```

### MIP

```python
df = pd.read_csv("../log/results_mip.csv")
df.columns
```

```python
df_flow = df[df["file"].str.contains("flow")]
df_job = df[df["file"].str.contains("job")]
df_open = df[df["file"].str.contains("open")]

print(len(df_flow), len(df_job), len(df_open))
print("Result Status Counts (flow):", df_flow['result_status'].value_counts().to_dict())
print("Result Status Counts (job):", df_job['result_status'].value_counts().to_dict())
print("Result Status Counts (open):", df_open['result_status'].value_counts().to_dict())
```

```python
df_flow[df_flow["result_status"] == 1]\
    .groupby(["file", "solver"])[['result_objective', 'result_runtime']]\
    .mean()\
    .reset_index()\
    .sort_values(by=["solver", "file"])
```

```python
df_job[df_job["result_status"] == 1]\
    .groupby(["file", "solver"])[['result_objective', 'result_runtime']]\
    .mean()\
    .reset_index()\
    .sort_values(by=["solver", "file"])
```

```python
df_open[df_open["result_status"] == 1]\
    .groupby(["file", "solver"])[['result_objective', 'result_runtime']]\
    .mean()\
    .reset_index()\
    .sort_values(by=["solver", "file"])
```
